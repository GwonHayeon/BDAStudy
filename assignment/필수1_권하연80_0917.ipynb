### sklearn의 공식 라이브러리

```python
import numpy as np
from sklearn.metrics import f1_score
```

```python
y_true = [0,1,2,0,1,2]
y_pred = [0,2,1,0,0,1]

f1_score_macro = f1_score(y_true, y_pred, average='macro')
f1_score_micro = f1_score(y_true, y_pred, average='micro')
f1_score_weighted = f1_score(y_true, y_pred, average='weighted')
f1_score_None = f1_score(y_true, y_pred, average=None)
print(f"macro : {f1_score_macro:.2f}")
print(f"micro : {f1_score_micro:.2f}")
print(f"weighted : {f1_score_weighted:.2f}")
print(f"None :",f1_score_None)
```

- macro : 산술평균
- micro : 총 참양성, 거짓음성, 거짓양성을 계산하여 전역적으로 측정항목을 계산
- weighted : 각 라벨의 측정항목을 계산하고 지원(각 라벨의 실제 인스턴스 수)을 기준으로 가중치가 부여된 평균
